```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('job-market.csv')
data.head()
df = pd.DataFrame(data)
df
```

```python
%matplotlib inline
```

```python
df['Count'] = df.groupby('Location')['Location'].transform('count')
nj = df.loc[:,("Location", "Count")]
nj
```

```python
jp = df.groupby('Location')['Area'].value_counts()
jp
```

```python
DataF = pd.DataFrame(nj)
DataF
```

```python
DataF = DataF.drop_duplicates(subset=['Location', 'Count'], keep='last')
```

```python
DataF = DataF.dropna(subset=['Location', 'Count'])
DataF
```

```python
X = list(DataF.iloc[:, 0])
Y = list(DataF.iloc[:, 1])
```

```python
plt.bar(X, Y, color='g')
plt.title("Number of Job Postings by Location")
plt.xlabel("Location")
plt.ylabel("Number of Job Postings")
plt.xticks(fontsize = 5)
plt.xticks(rotation=45, ha='right')
plt.show()
```
